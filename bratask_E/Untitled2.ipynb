```python
import os
import numpy as np

# Örnek dizin yolu (kendi dosya yolunuza göre güncelleyin)
dizin_yolu = "BraTask_ValidationData/images"

# Dizin içindeki dosyaları listeler
dosya_listesi = os.listdir(dizin_yolu)

# Her dosya için döngü
for dosya in dosya_listesi:
    # Numpy dosyası kontrolü
    if dosya.endswith(".npy"):
        # Dosya ismini yazdır
        print("Dosya İsmi:", dosya)
        
        # Numpy dosyasını yükle
        numpy_veri = np.load(os.path.join(dizin_yolu, dosya))
        
        # Veriyi yazdır
        print("Numpy Veri:", numpy_veri.shape)
```
